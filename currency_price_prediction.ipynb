```python
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse
import matplotlib.pyplot as plt

# Multiplicative Decomposition 
result_mul = seasonal_decompose(df['Close'], model='multiplicative', extrapolate_trend='freq')

# Additive Decomposition
result_add = seasonal_decompose(df['Close'], model='additive', extrapolate_trend='freq')

# Plot
plt.rcParams.update({'figure.figsize': (5,5)})
result_mul.plot().suptitle('Multiplicative Decompose', fontsize=22)
result_add.plot().suptitle('Additive Decompose', fontsize=22)
plt.show()
```

```python
#cheak the stationary
from statsmodels.tsa.stattools import adfuller
# ADF Test
#Augmented Dickeyâ€“Fuller test
result = adfuller(df.Close.values, autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
```

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from math import sqrt
```

```python
def ts_train_test_split(X,per=0.90):
    size = int(len(X) * per)
    train, test = X[0:size], X[size:len(X)]
    return train,test
```

```python
train,test = ts_train_test_split(df.loc['2003-12-23':,'RA'],per=0.90)
```

```python
train.head()
```

```python
len(train),len(test)
```

```python
train.index,test.index
```

```python
# Manually Configured ARIMA
model = ARIMA(train.values, order=(1,2,1)) # order is p,d,q
model_fit = model.fit( )
output = model_fit.predict(dynamic=False)
print(model_fit.summary())
```

```python
output_1 = model_fit.forecast(len(test))
#Root Mean Square Error
print("MSE  = ", mean_squared_error(test,output_1[0]))
print("Mean Square Log Error = ", mean_squared_log_error(test,output_1[0]))
print("RMSE  = ",sqrt(mean_squared_error(test,output_1[0])))
```

```python
print('expected=',test)
```

```python
#predict value
print('predicted=',output_1[0])
```

```python
#pridiction plot
#fig, ax = plt.subplots(figsize=(14,8))
ax.set_title("predicted price")
ax.set_xlabel('price')
ax.set_ylabel('date')
model_fit.plot_predict(len(test),ax=ax )
```

```python
# ACF and PACF (Auto corealation / paratial correlation)
import statsmodels.api as sm
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(model_fit.resid, lags=20, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(model_fit.resid, lags=20, ax=ax2)
```

```python
#fine tune by autoarima
import pmdarima as pm
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from math import sqrt
from pmdarima.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

y = df['Close']

train, test = train_test_split(y, train_size=4127)
```

```python
train.shape,test.shape
```

```python
test.head()
```

```python
#p is the order of the Auto Regressive (AR) term. It refers to the number of lags of Y to be used as predictors.
#d is  'Auto Regressiveâ€™ in ARIMA
#q is the order of the Moving Average (MA) term. 
model_autoARIMA =pm.auto_arima(train, start_p=0, start_q=0,
test='adf',       # use adftest to find optimal 'd'
max_p=3, max_q=3, # maximum p and q
m=1,              # frequency of series
d=None,           # let model determine 'd'
seasonal=True,   #  Seasonality 
error_action='ignore',  
suppress_warnings=True, 
stepwise=True)
print(model_autoARIMA.summary())
```

```python
model = ARIMA(train, order=(1, 1, 0))
fitted = model.fit(disp=-1)
print(fitted.summary())
```

```python
#fc = forcast ,stand error=se , confidence interval =conf
fc, se, conf = fitted.forecast( len(test), alpha=0.05)  # alpha =95% confidence
```

```python
fc.shape ,test.shape
```

```python
#Root Mean Square Error
print("MSE  = ", mean_squared_error(test,fc))
print("Mean Square Log Error = ", mean_squared_log_error(test,fc))
print("RMSE  = ",sqrt(mean_squared_error(test,fc)))
```

```python
import statsmodels.api as sm
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(fitted.resid, lags=20, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(fitted.resid, lags=20, ax=ax2)
```

```python
fc_series = pd.Series(fc, index=test.index)
lower_series = pd.Series(conf[:, 0], index=test.index)
upper_series = pd.Series(conf[:, 1], index=test.index)
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, color = 'blue', label='Actual  Price')
plt.plot(fc_series, color = 'orange',label='Predicted  Price')
plt.fill_between(lower_series.index, lower_series, upper_series, 
                 color='k', alpha=.10)
plt.title('Rupee Price Prediction')
plt.xlabel('Time')
plt.ylabel('Actual  Price')
plt.show()
```

# Apply SARIMA MODEL

```python
#resampleing model montly
y = df.Close.resample("M").mean()
train, test = train_test_split(y, train_size=183)
```

```python
y.shape,train.shape,test.shape,type(test)
```

```python
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from math import sqrt
model = sm.tsa.statespace.SARIMAX(train,order=(0,1,1),seasonal_order=(1,1,1,12))
result=model.fit()
# make your forecasts
forecasts_sarima= result.predict(start=184,end=212,dynamic=True) 
```

```python
forecasts_sarima.shape
```

```python
#Root Mean Square Error
print("MSE  = ", mean_squared_error(test,forecasts_sarima))
print("Mean Square Log Error = ", mean_squared_log_error(test,forecasts_sarima))
print("RMSE  = ",sqrt(mean_squared_error(test,forecasts_sarima)))
```

```python
result.plot_diagnostics(figsize=(10,8))
```

```python
# Visualize the forecasts (blue=train, green=forecasts)
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, color = 'blue', label='Actual  Price')
plt.plot(forecasts_sarima, color = 'red',label='Predicted  Price')
plt.title('Rupee Price Prediction (Montly)')
plt.xlabel('Time')
plt.ylabel('Actual  Price') 
plt.show()
```
